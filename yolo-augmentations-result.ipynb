```python
import os
import cv2
import matplotlib.pyplot as plt
from PIL import Image
from albumentations import Compose, HueSaturationValue, ShiftScaleRotate, Flip, OneOf, MotionBlur, MedianBlur, Blur

def visualize_augmentations(image_path, augmentations, save_path):
    # Load the original image
    image = cv2.imread(image_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Apply augmentations
    augmented_image = augmentations(image=image)['image']

    # Resize to (2700 x 1800)
    augmented_image = cv2.resize(augmented_image, (2700, 1800))

    # Save augmented image
    Image.fromarray(augmented_image).save(save_path)

# Define augmentations
augmentations = Compose([
    HueSaturationValue(
        hue_shift_limit=int(179 * 0.015),
        sat_shift_limit=int(255 * 0.7),
        val_shift_limit=int(255 * 0.4),
        p=1.0
    ),
    ShiftScaleRotate(
        shift_limit=0.1,
        scale_limit=0.5,
        rotate_limit=0,
        p=1.0
    ),
    Flip(p=0.5),
    OneOf([
        MotionBlur(p=0.5),
        MedianBlur(blur_limit=3, p=0.5),
        Blur(blur_limit=3, p=0.5),
    ], p=1.0),
], p=1.0)

# Path to the folder containing images
folder_path = "/kaggle/input/for-augmentation/for-Augmentation"

# Output folder for saving augmented images
output_folder = "/kaggle/working/"

# Iterate through each image in the folder
for filename in os.listdir(folder_path):
    if filename.endswith(".png"):
        image_path = os.path.join(folder_path, filename)
        
        # Iterate through three runs
        for run in range(1, 4):
            # Define the save path for the augmented image
            save_path = os.path.join(output_folder, f"{os.path.splitext(filename)[0]}_run_{run}.png")
            
            # Visualize augmentations and save the result
            visualize_augmentations(image_path, augmentations, save_path)
```

```python
import os
import cv2
import matplotlib.pyplot as plt
from PIL import Image
from albumentations import Compose, HueSaturationValue, ShiftScaleRotate, Flip, OneOf, MotionBlur, MedianBlur, Blur

def visualize_augmentations(image_path, augmentations, save_path):
    # Load the original image
    image = cv2.imread(image_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Apply augmentations
    augmented_image = augmentations(image=image)['image']

    # Resize to (2700 x 1800)
    augmented_image = cv2.resize(augmented_image, (2700, 1800))

    # Visualize original and augmented images
    plt.figure(figsize=(10, 5))

    plt.subplot(1, 2, 1)
    plt.imshow(image)
    plt.title("Original Image")

    plt.subplot(1, 2, 2)
    plt.imshow(augmented_image)
    plt.title("Augmented Image")

    plt.show()

    # Save augmented image
    Image.fromarray(augmented_image).save(save_path)

# Define augmentations
augmentations = Compose([
    HueSaturationValue(
        hue_shift_limit=int(179 * 0.015),
        sat_shift_limit=int(255 * 0.7),
        val_shift_limit=int(255 * 0.4),
        p=1.0
    ),
    ShiftScaleRotate(
        shift_limit=0.1,
        scale_limit=0.5,
        rotate_limit=0,
        p=1.0
    ),
    Flip(p=0.5),
    OneOf([
        MotionBlur(p=0.5),
        MedianBlur(blur_limit=3, p=0.5),
        Blur(blur_limit=3, p=0.5),
    ], p=1.0),
], p=1.0)

# Path to the folder containing images
folder_path = "/kaggle/input/for-augmentation/for-Augmentation"

# Output folder for saving augmented images
output_folder = "/kaggle/working/"

# Iterate through each image in the folder
for filename in os.listdir(folder_path):
    if filename.endswith(".png"):
        image_path = os.path.join(folder_path, filename)
        
        # Iterate through three runs
        for run in range(1, 4):
            # Define the save path for the augmented image
            save_path = os.path.join(output_folder, f"{os.path.splitext(filename)[0]}_run_{run}.png")
            
            # Visualize augmentations and save the result
            visualize_augmentations(image_path, augmentations, save_path)
```
